```python
import pandas as pd
```

```python
df = pd.read_excel("CLTV predict.xlsx")
```

```python
df.head()
```

```python
df.describe()
```

```python
order_value_df = df.groupby(['customer ID'])['order value'].mean().reset_index()
```

```python
order_value_df.head()
```

```python
frequency_df = df.groupby(['customer ID'])['date'].mean().reset_index()
```

```python
frequency_df.head()
```

```python
lifespan_df = df.groupby(['customer ID'])['month'].count().reset_index()
```

```python
lifespan_df.head()
```

```python
for i in range(len(lifespan_df)):
  lifespan_df['month'][i] = lifespan_df['month'][i] / 12
```

```python
lifespan_df.head()
```

```python
final_df = order_value_df.merge(frequency_df, on='customer ID').merge(lifespan_df, on='customer ID')
```

```python
final_df.head()
```

```python
final_df['cltv'] = None
for i in range(len(final_df)):
  final_df['cltv'][i] = final_df['order value'][i] * final_df['date'][i] * final_df['month'][i]
```
